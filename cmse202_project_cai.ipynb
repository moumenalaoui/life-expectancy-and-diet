vegetable oils vs life expectancy

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
life = pd.read_csv("life_expectancy.csv")
life.head()
```

```python
# plot the data
fig = plt.figure(figsize=(30, 10))
plt.plot(life["Country"], life["Life Expectancy   (both sexes)"], marker="o", color="blue", linewidth=3)

plt.title("Life Expectancy by Country")
plt.xlabel("Country")
plt.ylabel("Life Expectancy (both sexes)")


# rotate the x-axis labels for better visibility
plt.xticks(rotation=90)
plt.show()
```

```python
fat_sup = pd.read_csv("Fat_Supply_Quantity_Data.csv")

merged_data = pd.merge(life, fat_sup, on="Country")

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(merged_data["Vegetable Oils"], merged_data["Life Expectancy   (both sexes)"], alpha=0.5)
ax.set_xlabel("Vegetable Oil Consumption (kg/capita/yr)")
ax.set_ylabel("Life Expectancy")
ax.set_title("Life Expectancy vs Vegetable Oil Consumption by Country")
plt.savefig("Life Expectancy vs Vegetable Oil.jpg",dpi =1000)
plt.show()
```

```python
#assume comsuming more oils would shorten the life expectancy
oil_sort = fat_sup.sort_values("Vegetable Oils",ascending=False)
oil_sort.replace(np. nan,0)
oil_sort.head()
```

```python
df = fat_sup.merge(life,how = "right")
df.replace(np. nan,0)
```

```python
# Create a scatter plot
sns.scatterplot(data=df, x="Vegetable Oils", y="Life Expectancy   (both sexes)")

plt.title("Vegetable Oil Consumption vs. Life Expectancy")
plt.xlabel("Vegetable Oil Consumption")
plt.ylabel("Life Expectancy")

plt.show()
```

```python
asia_countries = ["Afghanistan", "Bahrain", "Bangladesh", "Bhutan", 
                  "Brunei", "Cambodia", "China", "India", "Indonesia",
                  "Iran", "Iraq", "Israel", "Japan", "Jordan", "Kazakhstan",
                  "Kuwait", "Kyrgyzstan", "Laos", "Lebanon", "Malaysia",
                  "Maldives", "Mongolia", "Myanmar", "Nepal",
                  "Korea, North","Korea, South" "Oman", "Pakistan", "Philippines",
                  "Qatar", "Russian Federation", "Saudi Arabia", "Singapore", "South Korea",
                  "Sri Lanka", "Syria", "Taiwan", "Tajikistan", "Thailand", 
                  "Timor-Leste", "Turkey", "Turkmenistan", "United Arab Emirates",
                  "Uzbekistan", "Vietnam", "Yemen"]
asia = df.loc[df["Country"].isin(asia_countries)]
sns.scatterplot(data=asia, x="Vegetable Oils", y="Life Expectancy   (both sexes)")

plt.title("Asia Vegetable Oil Consumption vs. Life Expectancy")
plt.xlabel("Vegetable Oil Consumption")
plt.ylabel("Life Expectancy")
```

```python
fig = plt.figure(figsize=(30, 10))
plt.xticks(rotation=90)
asia = asia.replace(np. nan,0)
plt.plot(asia["Country"],asia["Vegetable Oils"],color = "r",label = "Asia countrie VS Vegetable Oil Consumption")
plt.plot(asia["Country"],asia["Life Expectancy   (both sexes)"],color = "b",label = "Asia countrie VS Vegetable Oil Consumption")

plt.title("Asia Vegetable Oil Consumption & Life Expectancy")
plt.xlabel("Asia countrie")
plt.ylabel("Vegetable Oil Consumption                         &                          Life Expectancy")
plt.legend()
plt.savefig("Asia Vegetable Oil Consumption & Life Expectancy.jpg")
plt.show()
```
